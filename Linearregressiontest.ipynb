```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
from LinearRegression import LinearRegression
```

```python
X, y = datasets.make_regression(n_samples = 100, n_features = 1, random_state =4, noise = 20)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1233)
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], y, color="b", marker="o", s=30)
plt.show()
```

```python
reg = LinearRegression(lr=0.01)
```

```python
reg.fit(X_train, y_train)
```

```python
predictions = reg.predict(X_test)
```

```python
def mse(y_test, predictions):
    return np.mean((y_test - predictions) ** 2)

# Calculate MSE
mse_value = mse(y_test, predictions)
print("MSE:", mse_value)
```

```python
y_pred_line = reg.predict(X)
cmap = plt.get_cmap('ocean')
fig = plt.figure(figsize=(8,6))
m1 = plt.scatter(X_train, y_train, color=cmap(0.9), s=10)
m2 = plt.scatter(X_test, y_test, color=cmap(0.9), s=10)
plt.plot(X,y_pred_line, color='black', linewidth=2, label = 'predictions')
plt.show()
```
